```python
import sys
sys.path.append(r"C:\Users\ferna\OneDrive\Documentos\Disciplina_NLP\Atv_1")

import json
import glob
import os
import io
```

```python
pasta_json = r"/home/hurias/Documentos/Disciplina_NLP/Atv_1/data/corpus" 
count = 0

merge_texto = io.StringIO()
for filename in os.listdir(pasta_json):
    if filename.endswith('.json'):
        file_path = os.path.join(pasta_json, filename)
        try:
            with open(os.path.join(pasta_json, filename), 'r') as file:
                data = json.load(file)
                count += 1
                if 'text' in data:
                    merge_texto.write(data['text'] + " ")
                else:
                    print(f"A chave 'text' não encontrada no arquivo: {filename}")
        except json.JSONDecodeError:
            print(f"Erro ao decodificar o JSON no arquivo: {filename}")

print(f"Arquivos lidos: {count}")
```

```python
corpus = merge_texto.getvalue()
print(len(corpus))
```
